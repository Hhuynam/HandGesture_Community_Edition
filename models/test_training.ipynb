```python
import tensorflow as tf
import matplotlib.pyplot as plt

# c nh gc (dng TensorFlow)
image_path = r"D:\Project\HandGesture_Community_Edition\models\data2\test\fist\scene00001.png"
image = tf.io.read_file(image_path)
image = tf.image.decode_png(image, channels=3)  # c nh PNG, c 3 knh mu

# Resize nh v 150x150 v chun ha
resized_image = tf.image.resize(image, (150, 150)) / 255.0  # Scale v [0,1]

# Hin th nh gc v nh sau khi resize
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title("nh Gc")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(resized_image)
plt.title("nh Sau Khi Resize")
plt.axis("off")

plt.show()
```
